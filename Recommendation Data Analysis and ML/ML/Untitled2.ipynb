```python
weight = {1.5:['basketball court',
  'play court space with basketball backboard only',
  'temporary basketball court with backboard and ring',
  'basketball rings'],
         5.8:['concrete skate ramp', 'skate park','bmx track']}
```

```python
mel_all =pd.read_csv('mel_all.csv')
mel_all = mel_all.drop(columns=['Unnamed: 0', 'bar', 'balance', 'basic facilities', 'cool', 'rocker', 'swing', 'integrated', 'slide', 'see-saw', 'climb', 'mini-game', 'carousel', 'sand', 'walking', 'track', 'area'])
```

```python
mel_all
```

```python
equipment = list(mel_all['features'])
weights = [1] * len(equipment)
for i in equipment:
    index = equipment.index(i)
    temp = i.replace('.','').lower().split(', ')
    for j in temp:
        for k,v in weight.items():
            if j in v :
                weights[index] = k
            
```

```python
score = []
for i in range(len(weights)) :
    result = weights[i] * list(mel_all['total'])[i]
    score.append(result)
```

```python
for i in score :
    temp = round(i/17.4*99,2)
    score[score.index(i)] = temp
score
```

```python
mel_all['score']=score
```

```python
mel_all = mel_all.rename(columns={"name": "park_name", "features":"park_equipment", "score":"park_score"}).drop(columns=['total'])
```

```python
equipment = list(mel_all['park_equipment'])
for i in equipment:
    index = equipment.index(i)
    temp = i.replace('.','').lower().split(', ')
    equipment[index] = temp
mel_all['park_equipment'] = equipment
```
