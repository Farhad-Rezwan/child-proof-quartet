```python
def distance(lon1, lat1, lon2, lat2):
    dLat = radians(lat2 - lat1)
    dLon = radians(lon2 - lon1)
    lat1 = radians(lat1)
    lat2 = radians(lat2)
    r = 6378 * 1000
    a = sin(dLat/2)**2 + cos(lat1)*cos(lat2)*sin(dLon/2)**2
    c = 2*asin(sqrt(a))
    result = r*c
    return result
```

```python
p_lon = list(parks['LONGITUDE'])
p_lat = list(parks['LATITUDE'])
s_lon = list(sites['Longitude'])
s_lat = list(sites['Latitude'])
```

```python
distance_list = []
lowest_id = []
for index in range(len(list(parks['LONGITUDE']))):
    temp = None
    temp_id = None
    for si in range(len(list(sites['Site ID']))):
        d = distance(p_lon[index],p_lat[index],s_lon[si],s_lat[si])
        if temp == None or temp > d:
            temp = d
            temp_id = list(sites['Site ID'])[si]
    distance_list.append(temp)
    lowest_id.append(temp_id)
```

```python
parks['DISTANCE'] = distance_list
parks['SENSOR'] = lowest_id
```
